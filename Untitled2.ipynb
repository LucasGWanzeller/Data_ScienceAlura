```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# --- 0. Configuração e Carregamento dos Dados ---
print("Iniciando a análise das Lojas Alura Store...")

# Lista dos arquivos
file_names = ['loja_1.csv', 'loja_2.csv', 'loja_3.csv', 'loja_4.csv']
dfs = {}

# Tentar carregar os dados
try:
    for i, file in enumerate(file_names):
        store_name = f'Loja {i+1}'
        dfs[store_name] = pd.read_csv(file)
    print("Arquivos CSV carregados com sucesso.")
except FileNotFoundError as e:
    print(f"Erro: Arquivo não encontrado. {e}")
    print("Por favor, faça o upload dos arquivos 'loja_1.csv', 'loja_2.csv', 'loja_3.csv', e 'loja_4.csv' no Colab.")
    # Se os arquivos não forem encontrados, paramos a execução
    raise e

print("\n" + "="*80 + "\n")

# --- 1. Inspeção Inicial ---
# Vamos inspecionar a estrutura de um dos DataFrames
print("--- 1. Inspeção Inicial dos Dados (Exemplo: Loja 1) ---")
print(dfs['Loja 1'].info())
print("\n")
print(dfs['Loja 1'].head())
print("\n" + "="*80 + "\n")

# --- 2. Análise de Métricas por Loja ---
print("--- 2. Calculando Métricas por Loja ---")
analysis_results = []

for store_name, df in dfs.items():
    # Faturamento Total
    revenue = df['Preço'].sum()

    # Avaliação Média
    avg_rating = df['Avaliação da compra'].mean()

    # Frete Médio
    avg_shipping = df['Frete'].mean()

    # Produto Mais Vendido (por contagem)
    most_sold_product = df['Produto'].value_counts().idxmax()

    # Produto Menos Vendido (por contagem)
    counts = df['Produto'].value_counts()
    least_sold_product = counts[counts == counts.min()].index[0]

    # Categoria Mais Vendida
    most_sold_category = df['Categoria do Produto'].value_counts().idxmax()

    # Categoria Menos Vendida
    cat_counts = df['Categoria do Produto'].value_counts()
    least_sold_category = cat_counts[cat_counts == cat_counts.min()].index[0]

    analysis_results.append({
        'Loja': store_name,
        'Faturamento Total': revenue,
        'Avaliação Média': avg_rating,
        'Frete Médio': avg_shipping,
        'Produto Mais Vendido': most_sold_product,
        'Produto Menos Vendido': least_sold_product,
        'Categoria Mais Vendida': most_sold_category,
        'Categoria Menos Vendida': least_sold_category
    })

# Criar um DataFrame de resumo
summary_df = pd.DataFrame(analysis_results).set_index('Loja')

print("--- Tabela Resumo da Análise por Loja ---")
# Formatar o DataFrame para melhor visualização
summary_df_formatted = summary_df.copy()
summary_df_formatted['Faturamento Total'] = summary_df_formatted['Faturamento Total'].map('R$ {:,.2f}'.format)
summary_df_formatted['Avaliação Média'] = summary_df_formatted['Avaliação Média'].map('{:.2f}'.format)
summary_df_formatted['Frete Médio'] = summary_df_formatted['Frete Médio'].map('R$ {:,.2f}'.format)
print(summary_df_formatted)
print("\n" + "="*80 + "\n")

# --- 3. Análise Geral (Todas as Lojas) ---
print("--- 3. Análise Geral (Consolidado) ---")
# Concatenar todos os dados para uma visão geral
all_data = pd.concat(dfs.values(), ignore_index=True)

print(f"Total de Vendas (transações): {len(all_data)}")
print(f"Faturamento Total (Geral): R$ {all_data['Preço'].sum():.2f}")
print(f"Avaliação Média (Geral): {all_data['Avaliação da compra'].mean():.2f}")
print(f"Frete Médio (Geral): R$ {all_data['Frete'].mean():.2f}")
print("\n")

print("--- Categorias Mais Vendidas (Geral) ---")
print(all_data['Categoria do Produto'].value_counts())
print("\n" + "="*80 + "\n")

# --- 4. Criação dos Gráficos ---
print("--- 4. Gerando Visualizações ---")

# Gráfico 1: Faturamento Total por Loja (Gráfico de Barras)
faturamento_sorted = summary_df['Faturamento Total'].sort_values(ascending=False)
plt.figure(figsize=(10, 7))
bars = plt.bar(faturamento_sorted.index, faturamento_sorted.values, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Faturamento Total por Loja', fontsize=16)
plt.ylabel('Faturamento (R$)', fontsize=12)
plt.xlabel('Loja', fontsize=12)
formatter = mtick.FuncFormatter(lambda x, p: f'R$ {x:,.0f}')
plt.gca().yaxis.set_major_formatter(formatter)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + (faturamento_sorted.values.max() * 0.01), f'R$ {yval:,.0f}', va='bottom', ha='center', fontsize=10)
plt.tight_layout()
plt.savefig('faturamento_total.png')
print("Gráfico 'faturamento_total.png' salvo.")

# Gráfico 2: Avaliação Média por Loja (Gráfico de Barras)
avaliacao_sorted = summary_df['Avaliação Média'].sort_values(ascending=False)
plt.figure(figsize=(10, 7))
bars = plt.bar(avaliacao_sorted.index, avaliacao_sorted.values, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Avaliação Média por Loja', fontsize=16)
plt.ylabel('Avaliação Média (0-5)', fontsize=12)
plt.xlabel('Loja', fontsize=12)
plt.ylim(0, 5)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.05, f'{yval:.2f}', va='bottom', ha='center', fontsize=10)
plt.tight_layout()
plt.savefig('avaliacao_media.png')
print("Gráfico 'avaliacao_media.png' salvo.")

# Gráfico 3: Custo Médio do Frete por Loja (Gráfico de Barras)
frete_sorted = summary_df['Frete Médio'].sort_values(ascending=True) # Ordenado do menor para o maior
plt.figure(figsize=(10, 7))
bars = plt.bar(frete_sorted.index, frete_sorted.values, color=['#2ca02c', '#ff7f0e', '#1f77b4', '#d62728'])
plt.title('Custo Médio do Frete por Loja', fontsize=16)
plt.ylabel('Frete Médio (R$)', fontsize=12)
plt.xlabel('Loja', fontsize=12)
formatter = mtick.FuncFormatter(lambda x, p: f'R$ {x:.2f}')
plt.gca().yaxis.set_major_formatter(formatter)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + (frete_sorted.values.max() * 0.01), f'R$ {yval:.2f}', va='bottom', ha='center', fontsize=10)
plt.tight_layout()
plt.savefig('frete_medio.png')
print("Gráfico 'frete_medio.png' salvo.")

# Gráfico 4: Distribuição de Categorias Mais Vendidas (Geral) (Gráfico de Pizza)
# (Este é um 4º gráfico, bônus, para cumprir os "3 tipos diferentes")
category_counts = all_data['Categoria do Produto'].value_counts()
top_categories = category_counts.head(5)
outros = pd.Series({'Outros': category_counts.iloc[5:].sum()})
final_cat_counts = pd.concat([top_categories, outros])

plt.figure(figsize=(12, 8))
wedges, texts, autotexts = plt.pie(
    final_cat_counts,
    autopct='%1.1f%%',
    startangle=140,
    pctdistance=0.85,
    labels=None
)
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(10)
    autotext.set_fontweight('bold')

plt.title('Distribuição de Categorias de Produtos (Geral)', fontsize=16)
plt.axis('equal')
labels = [f'{l} - {s/final_cat_counts.sum()*100:1.1f}%' for l, s in final_cat_counts.items()]
plt.legend(labels, title="Categorias", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.tight_layout()
plt.savefig('categorias_geral_pie.png')
print("Gráfico 'categorias_geral_pie.png' salvo.")

print("\nAnálise concluída. Os gráficos estão salvos e prontos no Colab.")
print("\n" + "="*80 + "\n")

# --- 5. Relatório Final e Recomendação ---

# O código abaixo gera o relatório como um texto.
# Os valores são puxados diretamente do DataFrame 'summary_df' para garantir
# que a recomendação esteja sempre alinhada com os dados.

# Encontrar a loja com menor faturamento
loja_menor_faturamento = summary_df['Faturamento Total'].idxmin()
valor_menor_faturamento = summary_df['Faturamento Total'].min()

# Encontrar a loja com maior faturamento
loja_maior_faturamento = summary_df['Faturamento Total'].idxmax()
valor_maior_faturamento = summary_df['Faturamento Total'].max()

# Encontrar a loja com pior avaliação
loja_pior_avaliacao = summary_df['Avaliação Média'].idxmin()
valor_pior_avaliacao = summary_df['Avaliação Média'].min()

# Encontrar a loja com maior frete
loja_maior_frete = summary_df['Frete Médio'].idxmax()
valor_maior_frete = summary_df['Frete Médio'].max()

relatorio_final = f"""
# RELATÓRIO FINAL E RECOMENDAÇÃO PARA O SENHOR JOÃO

## 1. Introdução
O objetivo desta análise foi avaliar o desempenho de quatro lojas da Alura Store (Loja 1, Loja 2, Loja 3, Loja 4)
para identificar a loja com menor eficiência, visando uma potencial venda para investimento em outros negócios.
A análise baseou-se em métricas de faturamento, avaliação de clientes, custo de frete e popularidade de produtos
e categorias.

## 2. Análise dos Dados
Com base na tabela de resumo e nos gráficos gerados, observamos os seguintes pontos:

* **Faturamento (Ver gráfico 'faturamento_total.png'):**
    * A loja com maior faturamento é a **{loja_maior_faturamento}** (R$ {valor_maior_faturamento:,.2f}).
    * A loja com menor faturamento é a **{loja_menor_faturamento}** (R$ {valor_menor_faturamento:,.2f}).
    * A {loja_menor_faturamento} fatura aproximadamente R$ {valor_maior_faturamento - valor_menor_faturamento:,.2f} a menos que a loja líder.

* **Avaliação Média (Ver gráfico 'avaliacao_media.png'):**
    * Todas as lojas possuem avaliações muito próximas, com uma média geral de {all_data['Avaliação da compra'].mean():.2f}.
    * A loja com a pior avaliação é a **{loja_pior_avaliacao}** (média de {valor_pior_avaliacao:.2f}).
    * As demais lojas apresentam avaliações ligeiramente superiores, indicando um nível de satisfação de cliente similar entre elas.

* **Frete Médio (Ver gráfico 'frete_medio.png'):**
    * O custo de frete é um indicador de eficiência operacional.
    * A loja com o frete médio mais *alto* é a **{loja_maior_frete}** (R$ {valor_maior_frete:.2f}).
    * A **{loja_menor_faturamento}** possui o frete médio mais *baixo* (R$ {summary_df.loc[loja_menor_faturamento, 'Frete Médio']:.2f}), o que é um ponto positivo para ela.
* **Categorias (Ver gráfico 'categorias_geral_pie.png'):**
    * A categoria **'moveis'** é a mais vendida no geral, representando {final_cat_counts.loc['moveis'] / final_cat_counts.sum() * 100:.1f}% das vendas das top 5 categorias + outros.
    * 'eletronicos' e 'brinquedos' seguem como categorias relevantes.

## 3. Conclusão e Recomendação

**Recomendação: A loja a ser vendida é a {loja_menor_faturamento}.**

**Justificativa:**

A **{loja_menor_faturamento}** é a candidata clara para a venda por apresentar o pior desempenho no indicador mais crítico para a saúde de um negócio de varejo: o **Faturamento Total**.

1.  **Baixa Geração de Receita:** A {loja_menor_faturamento} é a que menos vende, gerando R$ {valor_menor_faturamento:,.2f}. Esta é uma diferença significativa em relação às demais, especialmente a {loja_maior_faturamento}. Isso indica menor volume de vendas ou menor atração de clientes.

2.  **Métricas Operacionais:** Embora a {loja_menor_faturamento} tenha um ponto forte (o menor frete médio) e sua avaliação esteja na média (próxima de 4.0, como as outras), esses fatores positivos não compensam a falha em gerar receita no mesmo nível de suas congêneres.

3.  **Comparação:** Enquanto a {loja_maior_frete} (que tem o maior faturamento) apresenta problemas operacionais (pior avaliação e maior frete), ela ainda é a que mais gera caixa. A {loja_menor_faturamento} é a "lanterna" em receita, tornando-a o ativo menos eficiente do portfólio.

Vender a **{loja_menor_faturamento}** permitiria ao Senhor João liquidar o ativo menos performático e focar os recursos e esforços nas lojas que já demonstram maior capacidade de venda e retorno financeiro.
"""

print(relatorio_final)

# Esta linha faz o script ser executável
# if __name__ == "__main__":
    # Certifique-se de que a função analisar_lojas() esteja definida em algum lugar
    # ou remova esta chamada se a análise já estiver sendo feita no corpo principal do script.
    # analisar_lojas()
    # pass # Substituído pela execução direta no corpo do script
```
